```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
file = pd.read_csv('Social_Network_Ads.csv')
```

```python
file
```

```python
file.drop('User ID',inplace=True,axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
Label = LabelEncoder()
file.iloc[:,[0]] = Label.fit_transform(file.iloc[:,[0]])
file
```

```python
file.info()
```

```python
file.describe()
```

```python
file.corr()
```

```python
x = file.iloc[:,[1,2]].values
y = file.iloc[:,[-1]].values
```

```python
from sklearn.preprocessing import StandardScaler
Stand = StandardScaler()
Stand.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=True)
```

```python
print(len(x_train),len(x_test))
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(x_train,y_train)
```

```python
reg_predict = reg.predict(x_test)
```

```python
from sklearn.metrics import classification_report
cr = classification_report(reg_predict,y_test)
print(cr)
```

```python
#Navie Bayes Classifier
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
nb_pred = nb.predict(x_test)
```

```python
from sklearn.metrics import classification_report
cr_nb = classification_report(nb_pred,y_test)
print(cr_nb)
```

```python
# KNN 
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier(n_neighbors=5)
kn.fit(x_train,y_train)
```

```python
kn_pred = kn.predict(x_test)
```

```python
from sklearn.metrics import classification_report
cr_kn = classification_report(kn_pred,y_test)
print(cr_kn)
```

```python
#Support Vector Machine
from sklearn.svm import SVC
svm_linear = SVC(kernel='linear')
svm_linear.fit(x_train,y_train)
```

```python
svm_lin_pred = svm_linear.predict(x_test)
```

```python
from sklearn.metrics import classification_report
cr_linear = classification_report(svm_lin_pred,y_test)
print(cr_linear)
```

```python
from sklearn.svm import SVC
svm_sig = SVC(kernel='sigmoid')
svm_sig.fit(x_train,y_train)
```

```python
svm_sig_pred = svm_sig.predict(x_test)
```

```python
from sklearn.metrics import classification_report
cr_sig = classification_report(svm_sig_pred,y_test)
print(cr_sig)
```

```python
from sklearn.svm import SVC
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(x_train,y_train)
```

```python
svm_rbf_pred = svm_rbf.predict(x_test)
```

```python
from sklearn.metrics import classification_report
cr_rbf = classification_report(svm_rbf_pred,y_test)
print(cr_rbf)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
Dt = DecisionTreeClassifier(criterion='gini')
Dt.fit(x_train,y_train)
```

```python
pred_Dt = Dt.predict(x_test)
```

```python
from sklearn.metrics import classification_report
cr_Dt = classification_report(pred_Dt,y_test)
print(cr_Dt)
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 5,criterion = 'gini')
rf.fit(x_train,y_train)
```

```python
pred_rf = rf.predict(x_test)
```

```python
from sklearn.metrics import classification_report
cr_rf = classification_report(pred_rf,y_test)
print(cr_rf)
```

```python
# Accuracy score of all the Algorithms
print('Logisitic_Regression accuracy :',reg.score(x_test,y_test)*100)
print('Navie_Bayes_accuracy :',nb.score(x_test,y_test)*100)
print('K_nearest_negihbors_accuracy :',kn.score(x_test,y_test)*100)
print('Support_vector_machine_linear :',svm_linear.score(x_test,y_test)*100)
print('Support_vector_machine_sigmoid :',svm_sig.score(x_test,y_test)*100)
print('Support_vector_machine_rbf :',svm_rbf.score(x_test,y_test)*100)
print('Decision_tree :',Dt.score(x_test,y_test)*100)
print('Random_forest_classifier :',rf.score(x_test,y_test)*100)
```

### Therefore the Random_forest_classifier has the maximum accuracy.
